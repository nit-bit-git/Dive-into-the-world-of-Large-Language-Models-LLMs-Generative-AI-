```python
import os
import tarfile
import io
from tqdm import tqdm
import chardet
import lzma

def decode_text(byte_content):
    try:
        return byte_content.decode('utf-8')
    except UnicodeDecodeError:
        detected = chardet.detect(byte_content)
        encoding = detected['encoding']
        try:
            return byte_content.decode(encoding)
        except:
            return byte_content.decode('latin-1')

def process_xz_file(xz_file_path):
    print(f"Processing XZ file: {xz_file_path}")
    with lzma.open(xz_file_path, "rb") as xz_file:
        content = xz_file.read()
        print(f"Read {len(content)} bytes from XZ file")
        return content

folder_path = "C:/Users/nitis/Desktop/LLM/openwebtext/subsets"
output_file_train = "train_split{}.txt"
output_file_val = "val_split{}.txt"
vocab_file = "vocab.txt"

# split_file = int(input("How many files do you want to split this into? "))
files = [f for f in os.listdir(folder_path) if f.endswith('.tar')]
total_files = len(files)
split_index = int(total_files * 0.9) # 90% for training 
files_train = files[:split_index]
files_val = files[split_index:]
print(f"Total files: {total_files}")

# max_count = total_files // split_file if split_file != 0 else total_files
vocab = set()


# print(f"Processing split {i+1}/{split_file}")
with open(output_file_train, "w", encoding="utf-8") as outfile:
        for filename in tqdm(files_train, total=len(files_train)):
            file_path = os.path.join(folder_path, filename)
            print(f"\nProcessing file: {file_path}")
            try:
                with tarfile.open(file_path, "r") as outer_tar:
                    # print(f"Opened outer tar: {filename}")
                    for member in outer_tar.getmembers():
                        # print(f"Found member: {member.name}")
                        if member.name.endswith('.xz'):
                            print(f"Found XZ file: {member.name}")
                            f = outer_tar.extractfile(member)
                            if f is not None:
                                xz_content = f.read()
                                with lzma.open(io.BytesIO(xz_content), "rb") as xz_file:
                                    byte_content = xz_file.read()
                                    text = decode_text(byte_content)
                                    print(f"Decoded text length: {len(text)}")
                                    outfile.write(text)
                                    characters = set(text)
                                    vocab.update(characters)
                                    print(f"Updated vocab. Current size: {len(vocab)}")
            except Exception as e:
                print(f"Error processing {filename}: {str(e)}")
        # files = files[max_count:]
with open(output_file_val, "w", encoding="utf-8") as outfile:
        for filename in tqdm(files_val, total=len(files_val)):
            file_path = os.path.join(folder_path, filename)
            print(f"\nProcessing file: {file_path}")
            try:
                with tarfile.open(file_path, "r") as outer_tar:
                    # print(f"Opened outer tar: {filename}")
                    for member in outer_tar.getmembers():
                        # print(f"Found member: {member.name}")
                        if member.name.endswith('.xz'):
                            # print(f"Found XZ file: {member.name}")
                            f = outer_tar.extractfile(member)
                            if f is not None:
                                xz_content = f.read()
                                with lzma.open(io.BytesIO(xz_content), "rb") as xz_file:
                                    byte_content = xz_file.read()
                                    text = decode_text(byte_content)
                                    print(f"Decoded text length: {len(text)}")
                                    outfile.write(text)
                                    characters = set(text)
                                    vocab.update(characters)
                                    print(f"Updated vocab. Current size: {len(vocab)}")
            except Exception as e:
                print(f"Error processing {filename}: {str(e)}")
        # files = files[max_count:]
# print(f"Finished processing split {i+1}. Checking file size...")
file_size = os.path.getsize(output_file_train)
print(f"Size of {output_file_train}: {file_size} bytes")
file_size = os.path.getsize(output_file_val)
print(f"Size of {output_file_val}: {file_size} bytes")

print("Writing vocabulary file...")
with open(vocab_file, "w", encoding="utf-8") as vfile:
    for char in sorted(vocab):
        vfile.write(char + "\n")

print(f"Vocabulary size: {len(vocab)}")
print("Processing complete.")
```
